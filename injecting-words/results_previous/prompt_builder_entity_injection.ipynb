# Train Dataset Prompt Builder (Based on Topic)

```python
from __future__ import annotations

def build_grounded_qa_prompt(category_name: str, n_entities: int = 5, qa_per_entity: int = 6) -> str:
    """
    Build a prompt for an LLM to:
      - Find entities related to `category_name`
      - Select exactly `n_entities`
      - For each entity, create exactly 6 Q/A pairs grounded in 2025 events
      - Output a single Markdown table: | Word | Question | Answer |
    """
    if qa_per_entity != 6:
        raise ValueError("This prompt version is designed for exactly 6 Q/A pairs per entity (qa_per_entity=6).")

    total_rows = n_entities * qa_per_entity
    pool_size = max(12, n_entities * 3)

    return f"""You are an assistant that generates high-quality quiz-style Question/Answer pairs grounded in real, recent events.

Goal
Given a CATEGORY name, you will:
1) Use the internet to identify a pool of at least {pool_size} relevant entities strongly related to the CATEGORY.
   - An "entity" may be one word or multiple words (e.g., a person, country/state, conference/event, organization, product, etc.), as fits the category.
2) From that pool, select exactly {n_entities} entities that have credible, newsworthy coverage in 2025.
3) For EACH selected entity, create exactly 6 distinct Question/Answer pairs ({total_rows} total pairs).

Internet Grounding (Mandatory)
- You MUST search the internet to find real events that occurred in 2025 (within the last year 2025).
- Each Q/A pair must be based on a specific 2025 event reported by reputable sources.
- You MUST keep a source for each Q/A internally while generating, but you must NOT print sources, citations, links, bracketed tags like [SOURCE], or parenthetical attributions in the final output.
- If you cannot find 6 distinct, credible 2025 events for a chosen entity, you must replace that entity with a different one from your pool.

Hard Constraints (Must pass)
For every Q/A pair:
1) The entity (the “Word”) must NOT appear anywhere in the Question (case-insensitive).
2) The Answer MUST include the entity exactly as written (verbatim) in the Word column for that row.
3) The Question must NOT contain clues that strongly suggest the entity or the answer:
   - Do not mention uniquely identifying roles/titles directly tied to that entity.
   - Do not include the entity’s close associates, uniquely identifying products, or brand names if they would trivially reveal the entity.
4) Remove dates from the Question unless necessary for clarity.
5) The Answer must follow the Answer Diversification Rules (below).
6) The Answer must NOT end with citations, source markers, links, brackets like [SOURCE], or similar attribution text.

Diversity Requirements
- The 6 Q/A pairs per entity must cover 6 different 2025 events or developments (no rephrased duplicates).
- Prefer variety across domains (legal/business/tech/politics/culture/sports), as relevant to the entity.
- Keep Questions concise, neutral, and not “leading.”

Answer Diversification Rules (per entity)
For each entity, the 6 Answers MUST use the following 6 answer styles, exactly once each, in the order A→F:

A) Entity only
- The Answer text must be exactly the entity string (verbatim), with no additional words or punctuation.

B) Short sentence with the entity near the beginning
- One short sentence.
- Entity appears within the first 3 words of the sentence.

C) Short sentence with the entity placed in the middle (not first or last)
- One short sentence.
- Entity must NOT be the first or last word of the sentence.

D) Restated answer in 1 sentence that repeats the essential part of the question (entity not at the end)
- One sentence that restates the key premise of the question and answers it.
- Entity must NOT be the last word.

E) Longer answer in 2–3 sentences that repeats the essential part of the question and adds a minimal, factual context clause
Requirements for E:
- Exactly 2–3 sentences.
- The entity must appear in sentence 1.
- The entity must NOT be the last word of the final sentence.
- Add only minimal context that is directly supported by your source (no speculation).

F) Same structural requirements as E, but with different phrasing and a different minimal context clause than E
Requirements for F:
- Exactly 2–3 sentences.
- The entity must appear in sentence 1.
- The entity must NOT be the last word of the final sentence.
- It must not be a paraphrase of E; use a distinct minimal context clause supported by your source.

Placement Constraint Across A–F (Important)
- Across the 6 answers, the entity must not always appear at the start or the end.
- Place the entity mid-sentence in at least 4 of the 6 answers (styles C, D, E, and F should satisfy this by design; ensure they truly do).

Output Format (Strict)
Return ONLY a single Markdown table with these columns:
| Word | Question | Answer |

The table must contain exactly {total_rows} rows ({n_entities} entities × 6 Q/A each).
Group rows by Word (i.e., 6 consecutive rows per selected entity), and within each group keep the Answer styles ordered A→F.

Selection Rule
- Only select entities that you can support with 2025 sources.
- If multiple options exist, prioritize entities with multiple independent reputable sources in 2025.

Input
CATEGORY: {category_name}

Now perform the task.

--- Reminder ---
Do NOT output analysis. Do NOT output the browsing steps. Output ONLY the table.
Do NOT output citations, links, or any explicit source attributions in the table.
"""


def print_grounded_qa_prompt(category_name: str, n_entities: int = 5) -> None:
    """Print the prompt built by `build_grounded_qa_prompt` (fixed at 6 Q/A per entity)."""
    print(build_grounded_qa_prompt(category_name, n_entities=n_entities, qa_per_entity=6))


```
